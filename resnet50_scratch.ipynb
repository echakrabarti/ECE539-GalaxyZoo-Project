```python
from preprocessing import get_datasets
from model_helper import *
from keras.applications.resnet_v2 import ResNet50V2
from tensorflow.keras.applications.resnet_v2 import preprocess_input
from keras.models import Model, Sequential
from keras.layers import Dense, GlobalAveragePooling2D, AveragePooling2D, Flatten, Dropout, Input
import matplotlib.pyplot as plt
```

```python
train_dataset, validation_dataset, test_dataset = get_datasets(32)
```

```python
base_model = ResNet50V2(weights='imagenet', input_shape=(69, 69, 3), include_top=False)
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(2048, activation='relu')(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.2)(x)
predictions = Dense(37, activation='sigmoid')(x)
model = Model(inputs=base_model.input, outputs=predictions)

model.summary()
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
name = 'ResNet50 Scratch'
history = compile_and_fit(model, train_dataset, validation_dataset, 10)
plot_history(history, name)
save(model, test_dataset, name)
```
